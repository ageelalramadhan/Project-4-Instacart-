# 4.5 IC Combining & Exporting Data

In this notebook, I combine the cleaned Instacart orders data with the `orders_products_prior` data and export the result as a pickle file.

```python
# 01. Importing libraries and data

import pandas as pd
import os

# Set path to main project folder
path = r"/Users/a/Documents/Instacart Basket Analysis"

# Paths to input data
ords_cleaned_path = os.path.join(path, "02 Data", "Prepared Data", "orders_cleaned.csv")
ords_prior_path = os.path.join(path, "02 Data", "Original Data", "orders_products_prior.csv")

# Import cleaned orders data
df_ords = pd.read_csv(ords_cleaned_path)

# Import orders_products_prior data
df_ords_prior = pd.read_csv(ords_prior_path)
```

```python
# Quick preview and shape checks

df_ords.head(), df_ords_prior.head()
```

```python
df_ords.shape, df_ords_prior.shape
```

```python
# 02. Merge orders data with orders_products_prior

df_ords_products_combined = df_ords.merge(df_ords_prior, on="order_id", how="inner")

# Quick checks
df_ords_products_combined.head()
```

```python
df_ords_products_combined.shape
```

```python
# 03. Export merged dataframe as pickle

export_path = os.path.join(path, "02 Data", "Prepared Data", "orders_products_combined.pkl")
df_ords_products_combined.to_pickle(export_path)

export_path
```
