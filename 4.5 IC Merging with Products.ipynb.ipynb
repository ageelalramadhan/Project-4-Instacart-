# 4.5 – Merging orders_products_combined with Cleaned Products Data

```python
# 01. Import libraries
import pandas as pd
import os

# Set base path
path = r"/Users/a/Documents/Instacart Basket Analysis"

# Path to the pickle file
pickle_path = os.path.join(path, "02 Data", "Prepared Data", "orders_products_combined.pkl")

# Import the dataframe
df_orders_products = pd.read_pickle(pickle_path)

# Quick check
df_orders_products.head()
```

```python
df_orders_products.shape
```

```python
# 02. Import cleaned products data
prods_path = os.path.join(path, "02 Data", "Prepared Data", "products_cleaned.csv")

df_prods = pd.read_csv(prods_path)

# Quick check
df_prods.head(), df_prods.shape
```

```python
# 03. Merge orders_products with products
df_ords_prods_merge = df_orders_products.merge(
    df_prods,
    on = "product_id",
    how = "left",
    indicator = True
)

# Quick check
df_ords_prods_merge.head(), df_ords_prods_merge.shape
```

```python
df_ords_prods_merge["_merge"].value_counts()
```

```python
# Export final merged dataframe as pickle (best format for large files)
export_path = os.path.join(path, "02 Data", "Prepared Data", "ords_prods_merge.pkl")
df_ords_prods_merge.to_pickle(export_path)

export_path
```
